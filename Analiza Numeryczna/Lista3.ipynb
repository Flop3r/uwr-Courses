```python
import numpy as np
import matplotlib.pyplot as plt

# Funkcja do tworzenia wykresu porównawczego
def plot_results(x_vals, old_vals, new_vals, title):
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, old_vals, label='Old Formula', color='blue')
    plt.plot(x_vals, new_vals, label='New Formula', color='red', linestyle='--')
    plt.xlabel('x')
    plt.ylabel('Value')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def old_a(x):
    return 1 / (x**5 + np.sqrt(x**10 + 2024))

def new_a(x):
    if x >= 0:
        return 1 / (x**5 + np.sqrt(x**10 + 2024))
    else:
        return (np.sqrt(x**10 + 2024) - x**5) / 2024

# Generowanie wartości x oraz wartości funkcji
x_values = np.linspace(-100, 100, 500)
old_vals = old_a(x_values)
new_vals = np.array([new_a(x) for x in x_values])

# Rysowanie wykresu porównawczego
plot_results(x_values, old_vals, new_vals, "Zad1 a)")
```

```python
def old_b(x):
    return 10**8 * (np.e**x - np.e**(2*x))

def new_b(x):
    i = np.arange(1, 11)
    terms = np.sum(x**i / np.cumprod(i))  
    return -10**8 * np.e**x * terms    

x_values = np.linspace(-100, 100, 500)
old_vals = old_b(x_values)
new_vals = np.array([new_b(x) for x in x_values])

plot_results(x_values, old_vals, new_vals, "Zad1 b)")
```

```python
def old_c(x):
    return 6*x**(-3)*(np.arcsin(x) - x)

def new_c(x):
    return 1 + (9 / 20) * x**2 + (15 / 56) * x**3

x_values = np.linspace(-0.001, 0.001, 500)
old_vals = old_c(x_values)
new_vals = np.array([new_c(x) for x in x_values])

plot_results(x_values, old_vals, new_vals, "Zad1 c)")
```

```python
def old_d(x):
    return 4 * np.cos(x)**2 - 1

def factorial(n):
    return np.array([np.math.factorial(i) for i in n])

def new_d(x):
    n = np.arange(1, 11)
    factorials = factorial(2 * n)
    terms = (2**(2*n + 1)) * (-1)**n * x**(2*n)/ factorials
    return np.sum(terms)

x_values = np.linspace(-5, 5, 10000)
old_vals = old_d(x_values)
new_vals = np.array([new_d(x) for x in x_values])

plot_results(x_values, old_vals, new_vals, "Zad1 c)")
```

```python
def old_2(b):
    a = 1
    c = -1
    delta = b**2 - 4 * a * c
    x1 = (-b + np.sqrt(delta)) / (2 * a)
    x2 = (-b - np.sqrt(delta)) / (2 * a)
    return x1, x2

def new_2(b):
    a = 1
    c = 1
    delta = b**2 - 4 * a * c
    x1 = np.where(b >= 0, (-b + np.sqrt(delta)) / (2 * a), c / ((-b - np.sqrt(delta)) / (2 * a)))
    x2 = np.where(b >= 0, c / x1, (-b - np.sqrt(delta)) / (2 * a))
    return x1, x2

b_values = np.linspace(-5, 5, 500)

old_x1_vals, old_x2_vals = old_2(b_values)
new_x1_vals, new_x2_vals = new_2(b_values)

plt.figure(figsize=(10, 5))
plt.plot(b_values, old_x1_vals, label='Old x1', color='blue', linestyle='-')
plt.plot(b_values, old_x2_vals, label='Old x2', color='green', linestyle='-')
plt.plot(b_values, new_x1_vals, label='New x1', color='red', linestyle='--')
plt.plot(b_values, new_x2_vals, label='New x2', color='orange', linestyle='--')

plt.xlabel('b')
plt.ylabel('x')
plt.title("Porównanie pierwiastków x1, x2 dla starych i nowych wyników")
plt.legend()
plt.grid(True)
plt.show()
```
