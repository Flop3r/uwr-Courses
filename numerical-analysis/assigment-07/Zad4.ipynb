```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial(x, nodes):
    p = 1
    for node in nodes:
        p *= (x - node)
    return p

n_values = range(4, 21)
x = np.linspace(-1, 1, 400)

cols = 5
rows = -(-len(n_values) // cols)
```

```python
fig, axs = plt.subplots(rows, cols, figsize=(20, 10), sharex=True, sharey=False)
axs = axs.flatten()

for idx, n in enumerate(n_values):
    equidistant_nodes = np.linspace(-1, 1, n + 1)
    chebyshev_nodes = np.cos((2 * np.arange(n + 1) + 1) * np.pi / (2 * (n + 1)))

    p_equidistant = polynomial(x, equidistant_nodes)
    p_chebyshev = polynomial(x, chebyshev_nodes)
    
    ax = axs[idx]
    ax.plot(x, p_equidistant, label="Wzy rwnoodlege")
    ax.plot(x, p_chebyshev, label="Wzy Chebysheva")
    ax.set_title(f'n={n}')
    ax.legend()
    

    ymin = min(p_equidistant.min(), p_chebyshev.min())
    ymax = max(p_equidistant.max(), p_chebyshev.max())
    margin = 0.1 * (ymax - ymin)
    ax.set_ylim(ymin - margin, ymax + margin)

for ax in axs[len(n_values):]:
    ax.axis('off')

plt.suptitle("Comparison of Polynomials for Equidistant and Chebyshev Nodes")
plt.tight_layout()
plt.show()
```
